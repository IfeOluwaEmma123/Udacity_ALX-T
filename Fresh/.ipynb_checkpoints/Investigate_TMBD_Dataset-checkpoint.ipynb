### QUESTION 2: What is the REVENUE generated Year by Year?

It will be essential to firstly build a cut_into_quantile function to have the data in four quartile: 'Little', 'Average', 'Much', 'Very Much'
This will help to section the revenue range properly

```python
#Building the cut_into_quantile functions
def cut_into_quantile(dfname ,column_name):
# Defining the maximum and minimum values
    min_value = dfname[column_name].min()
    first_quantile = dfname[column_name].describe()[4]
    second_quantile = dfname[column_name].describe()[5]
    third_quantile = dfname[column_name].describe()[6]
    max_value = dfname[column_name].max()
# Defining the quartile alias.
    bin_edges = [ min_value, first_quantile, second_quantile, third_quantile, max_value]
# Assigning alias names
    bin_names = [ 'Little', 'Average', 'Much', 'Very Much'] 
# Assigning the revenue_range column name
    name = '{}_range'.format(column_name)
    dfname[name] = pd.cut(dfname[column_name], bin_edges, labels=bin_names, include_lowest = True)
    return dfname
```

```python
#Sorting for the most recent 5 years
dfyear =[2011,2012,2013,2014,2015]
df_q2 = pd.DataFrame()

#Applying quartiles to each year
for year in dfyear:
    dfn = df.query('release_year == "%s"' % year)
    dfn2 = cut_into_quantile(dfn,'revenue')
    df_q2 = df_q2.append(dfn2)
df_q2.info()
```

The above tables shows that revenue_range has been added to the table, Great!

```python
# The revenue_range and release_year are both used in grouping the data
dfq2_summary = df_q2.groupby(['release_year','revenue_range']).median()
#Showing the first 10 rows of the grouped table
dfq2_summary.tail(10)
```

The grouped table is visualized below:

```python
# Setting the plot positions
pos = list(range(len(dfq2_summary.query('revenue_range =="Little"'))))
width = 0.2 

#Setting the bars and axis
fig, ax = plt.subplots(figsize=(10,5))
plt.bar(pos,
        dfq2_summary.query('revenue_range =="Little"')['popularity'], 
        width,
        alpha=0.5,
        color='#273746',
        label= 'Little') 

plt.bar([p + width for p in pos], 
        dfq2_summary.query('revenue_range =="Average"')['popularity'],
        width, 
        alpha=0.5, 
        color='#B03A2E', 
        label='Average') 

plt.bar([p + width*2 for p in pos], 
        dfq2_summary.query('revenue_range =="Much"')['popularity'], 
        width, 
        alpha=0.5, 
        color='#27AE60', 
        label='Much') 

plt.bar([p + width*3 for p in pos], 
        dfq2_summary.query('revenue_range =="Very Much"')['popularity'], 
        width, 
        alpha=0.5, 
        color='#6C3483', 
        label='Very Much')

#Defining the labels for the x and y axes
ax.set_ylabel('popularity')
ax.set_xlabel('Revenue_Range by Year')
ax.set_title('2015 - 2011 Revenue_Range by Popularity')

#Applying the xticks to avoid a sqooshed plot
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels([2011,2012,2013,2014,2015])

#Defining the chart legend
plt.legend( loc='upper left')
plt.grid()
plt.show()
```

#### Conclusions from Question 2
<li>The chart showed that, the higher the movie popularity, the higher the revenue range.</li>
<li>The highest movie popularity was in 2015, where the highest revenue_range was also noticed.</li>
<li>The chart implies that, revenue_range has a significant impact on the movie popularity.</li>

### QUESTION 3: What is the Average VOTE SCORED by movies Year after Year?

To explore the distribution of revenue in different score rating levels (between 2011 - 2015).

```python
# The revenue_range and release_year are both used in grouping the data
dfq2_summary = df_q2.groupby(['release_year','revenue_range']).mean()
#Showing the first 10 rows of the grouped table
dfq2_summary.tail(10)
```

Visualizations will make the question more answerable.

```python
# Setting the plot positions
pos = list(range(len(dfq2_summary.query('revenue_range =="Little"'))))
width = 0.2 

#Setting the chart range
fig, ax = plt.subplots(figsize=(12,3))

#Setting the Chart parameters
plt.bar(pos,
        dfq2_summary.query('revenue_range =="Little"')['vote_average'],
        width,
        alpha=0.5, 
        color='#DE3163', 
        label= 'Low') 

plt.bar([p + width for p in pos],
        dfq2_summary.query('revenue_range =="Average"')['vote_average'],
        width, 
        alpha=0.5, 
        color='#6495ED', 
        label='Medium') 

plt.bar([p + width*2 for p in pos], 
        dfq2_summary.query('revenue_range =="Much"')['vote_average'], 
        width, 
        alpha=0.5, 
        color='#DFFF00', 
        label='Much') 

plt.bar([p + width*3 for p in pos],
        dfq2_summary.query('revenue_range =="Very Much"')['vote_average'], 
        width, 
        alpha=0.5, 
        color='#1C2833', 
        label='Very Much')

#Defining the labels for the x and y axes
ax.set_ylabel('vote average')
ax.set_xlabel('Revenue Range by Year')
ax.set_title('2011 - 2015 Revenue Range by Vote Average Score')

#Applying the xticks to avoid a sqooshed plot
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels([2015,2014,2013,2012,2011])

plt.ylim(2, 10)

#Defining the chart legend
plt.legend(loc='upper center')
plt.grid()
plt.show()
```

#### Conclusions from Question 3
<li>The chart revealed that, the higher the movie vote_score, the higher the revenue range.</li>
<li>The highest movie vote_score was in 2013, where the highest revenue_range was also noticed.</li>
<li>The chart showed that, revenue_range has a significant relationship with movie vote_score.</li>
